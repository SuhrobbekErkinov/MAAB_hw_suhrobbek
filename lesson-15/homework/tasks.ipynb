### task 1


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x ** 2 - 4 * x + 4

x_values = np.linspace(10, -10, 400)
y_values = f(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label=r'$f(x) = x^2 - 4x + 4$', color = 'blue', linewidth=2)
plt.xlabel("x-axis")  # Label for x-axis
plt.ylabel("y-axis")  # Label for y-axis
plt.title("Plot of $f(x) = x^2 - 4x + 4$")  # Title of the plot
plt.axhline(0, color='black', linewidth=1, linestyle='--')  # x-axis line
plt.axvline(0, color='black', linewidth=1, linestyle='--')  # y-axis line
plt.legend()  # Show legend
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid for readability

# Show the plot
plt.show()
```

### task 2

```python
x_values = np.linspace(0, 2*np.pi, 400)
sin_values = np.sin(x_values)
cos_values = np.cos(x_values)

plt.figure(figsize=(8, 5))
plt.plot(x_values, sin_values, label=r'$f(x) = sin(x)$', color='red', 
         linestyle='-', marker='o', markevery=40)
plt.plot(x_values, cos_values, label=r'$f(x) = cos(x)$', color='blue',
         linestyle='--', marker='s', markevery=40)

plt.xlabel("x-axis (radians)")
plt.ylabel("y-axis")
plt.title("Sine and Cosine functions")

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

### task 3

```python
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(0, 2*np.pi, 400)
x3 = np.linspace(-2, 2, 400)
x4 = np.linspace(0, 5, 400)

y1 = x1 ** 3
y2 = np.sin(x2)
y3 = np.exp(x3)
y4 = np.log(x4 + 1)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x1, y1, color='blue')
axes[0, 0].set_title(r'$f(x)=x^3$')
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("f(x)")

axes[0, 1].plot(x2, y2, color='red')
axes[0, 1].set_title(r'$f(x)=sin(x)$')
axes[0, 1].set_xlabel("x")
axes[0, 1].set_ylabel("f(x)")

axes[1, 0].plot(x3, y3, color='yellow')
axes[1, 0].set_title(r'$f(x)=e^x$')
axes[1, 0].set_xlabel("x")
axes[1, 0].set_ylabel("f(x)")

axes[1, 1].plot(x4, y4, color='green')
axes[1, 1].set_title(r'$f(x)=log(x+1)$')
axes[1, 1].set_xlabel("x")
axes[1, 1].set_ylabel("f(x)")

plt.tight_layout()

plt.show()
```

### task 4

```python
x_values = np.random.uniform(0, 10, 100)
y_values = np.random.uniform(0, 10, 100)

colors = np.random.rand(100)
sizes = np.random.randint(20, 100, 100)

plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values, s=sizes, c=colors, alpha=0.7, cmap='plasma', marker='o')

plt.title("Random Scatter plot of 100 points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar(label="Color Intensity")
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

### task 5

```python
data = np.random.normal(loc=0, scale=1, size=1000)
plt.figure(figsize=(8, 6))

plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.title("Histogram of Normally distributed data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

### task 6

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(X ** 2 + Y ** 2)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surf, ax=ax, shrink=0.6, aspect=10, label="Function Value")

ax.set_title("3D Surface Plot of f(x, y) = cos(x + y)")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("f(x, y)")

plt.show()
```

### task 7

```python
products = ["Product A", "Product B", "Product C", "Product D", "Product E"]
sales = [200, 150, 250, 175, 225]

colors = ["red", "green", "yellow", "blue", "orange"]

plt.figure(figsize=(8, 6))
plt.bar(products, sales, color=colors, edgecolor='black')

plt.title("Bar chart of random products")
plt.xlabel("Products")
plt.ylabel("Sales (in units)")
plt.ylim(0, max(sales) + 50)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
time_periods = ['T1', 'T2', 'T3', 'T4']
category_A = [5, 7, 6, 8]
category_B = [3, 5, 2, 6]
category_C = [4, 6, 5, 7]

plt.figure(figsize=(8, 6))  
plt.bar(time_periods, category_A, color='blue', label='Category A')
plt.bar(time_periods, category_B, color='green', bottom=category_A, label='Category B')
plt.bar(time_periods, category_C, color='red', bottom=np.array(category_A) + np.array(category_B), label='Category C')

plt.title("Stacked Bar Chart of Category Contributions Over Time")
plt.xlabel("Time Periods")
plt.ylabel("Values")
plt.legend()  
plt.grid(axis='y', linestyle="--", alpha=0.6)  

plt.show()
```
